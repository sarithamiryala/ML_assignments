# Building the model

```python
# linear regression model.
# it is a simple linear regression model i.e one dependent variable and one independent variable
# y = Ax1+Bx2+Cx3+Dx4+Ex5+F
# A,B,C,D,E is coefficient
# F is intercept
# aim is to find A,B,C,D,E and F so that we get mininum error
```

```python
#model building
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
# training the model or fitting the model

lr.fit(X_train,y_train)
```

```python
# testing the model
y_pred = lr.predict(X_test)
```

# Model Evaluation

```python
#evaluation metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
print(f'mean squared error is {mse}')
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
print(f'R2 score of the model is {r2}')
```

```python
print(f'intercept of the linear model id {lr.intercept_}')
```

```python
print(f'slope of the linear model id {lr.coef_}')
```
